```python
import matplotlib.pyplot as plt

def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
   
plot_correlation_map(newtit)   
```

this map allows to give us the correlation between all the columns by specifying us in each box the rate of correlation 
interpretation:

we are going to focus on the survival column we notice that it has a strong correlation with the class, the sex and the price these last results confirm our interpretation of before
